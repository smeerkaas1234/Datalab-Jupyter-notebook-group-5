# **Sprint 5**

## **Ethische aspecten** 

De informatie die wij hebben is op buurtniveau, wanneer deze informatie op persoonsniveau gebruikt wordt kan dit risico's met zich mee brengen. Deze risico's staan hier beneden, toegepast op basis van de 7 ethic guidelines van AI.

- Als eerste hebben we de regel dat AI persoongegevens moet beschermen, als deze informatie op basis van persoongegevens wordt gebruikt door bedrijven met AI, beschadigd dit het gevoel dat mensen hun informatie beschermd wordt omdat AI hun persoonlijke gegevens beheert en gebruikt. De risico's zijn dus dat mensen hun peroonslijke gegevens niet meer beschermd worden wegens het gebruik van AI.

- Hierna hebben we het risico dat persoongegevens data risico-voller is dan data op groter buurt niveau, bij fouten (technisch) of een datalek is er sprake van een grotere schade op gerichtere klanten waardoor AI de veiligheid niet kan garanderen.

- Een ander groot risico van het gebruik van persoongegevens door AI, is dat AI veel kan zeggen over hoe een persoon energie verbruikt/wanneer etc.Hierdoor kan AI patronen aanmaken die gevoelig liggen. Elektriciteitsbeheerders moeten meer privacy/datakwaliteitseisen opstellen op de AI modulle om ervoor te zorgen dat deze de privacy niet schendt van persoongegevens.

- Ook is transparantie een risico bij het gebruik van persoongegevens, de beheerder moet precies kunnen uitleggen welke persoonlijke data wordt verzameld en waarom. AI zal per persoon kijken en hiervoor systemen aanmaken. Dit kan ervoor zorgen dat klanten bang zijn voor misbruik van hun gegevens.

- Een ander risico is: Als er per persoon gekeken wordt, kunnen algoritmes onbewust bepaalde groepen mensen als 'slecht' zien. Omdat deze mensen bijvoorbeeld 'teveel' energie verbruiken. Met buurtdata is dat risico kleiner, omdat individuele verschillen minder zichtbaar zijn.

- Wanneer klanten het idee krijgen dat hun persoonlijke energiegedrag steeds wordt bewaard, kan dit het vertrouwen in duurzame innovaties verminderen. Dit kan afwegend werken op maatschappelijke steun.

- Het laatste risico is: Bij persoonsdata kunnen fouten direct gevolgen hebben voor menig individu, zoals verkeerde facturen of afsluitingen. De organisatie krijgt daardoor meer verantwoordelijkheid en kans op klachten of juridische problemen. Omdat deze als verantwoordelijk of aansprakelijk gezien kan worden.

# **Advies**
## **Inleiding**

Voor dit advies hebben we gekeken naar de CBS-data die we de afgelopen weken hebben geanalyseerd. We hadden data over meer dan `17.000` buurten verspreid over `355` gemeenten in heel Nederland.

Eerlijk gezegd hadden we van tevoren verwacht dat dingen zoals inkomen of leeftijd een grote rol zouden spelen bij hoeveel stroom mensen verbruiken. Maar dat bleek eigenlijk niet zo te zijn. Wat wél heel erg uitmaakte was het type woning waar mensen in wonen. In dit advies gaan we verder in op de grootste impact die dat heeft gehad.

---

## **Onze belangrijkste bevindingen**

### **Woningtype heeft verrassend veel invloed**

Toen we de correlaties berekenden tussen de verschillende woningtypes en energieverbruik kwamen we erachter dat twee-onder-een-kap woningen het meest energiezuinige zijn . De correlatie was `-0.35`, wat betekent dat gemeentes met meer van dit soort woningen gemiddeld minder energie verbruiken.

Bij appartementen was het andersom. We hadden gedacht dat die zuinig zouden zijn omdat ze kleiner zijn, maar de correlatie met gasverbruik was juist positief (r = 0.70). Waarom precies weten we niet zeker, maar het zou kunnen dat veel oudere flats gewoon slecht geisoleerd zijn.

Hier een overzichtje van wat we vonden:
- Twee-onder-een-kap: `r = -0.35` (zuinig)
- Tussenwoningen: `r = -0.26` (ook best zuinig)  
- Hoekwoningen: `r = -0.30`
- Vrijstaand: `r = 0.09` (klein hoger vebruik)
- Appartementen: `r = 0.70` met gasverbruik (veel hoger dan verwacht)

<br>

### **Gas en elektriciteit hangen samen**

Dit vonden we ook wel interessant: het gasverbruik en elektriciteitsverbruik correleren heel sterk met elkaar, namelijk `r = 0.81`. Dat is echt heel hoog. Het zou kunnen betekenen dat als je maatregelen neemt om gas te besparen, dat je dan automatisch ook elektriciteit bespaart. Misschien omdat beide afhankelijk zijn van isolatie ofzo.

<br>

### **Veel huizen zijn oud**

Van alle woningen in onze dataset is zo'n `83%` gebouwd voor het jaar 2000. Dat zijn dus huizen die al minstens 25 jaar oud zijn. Oudere huizen zijn vaak minder goed geisoleerd, dus dit verklaart wel een deel van het hoge verbruik.

---

## **Wat zou er gedaan kunnen worden?**

We hebben ons advies opgesplitst per type gebied, want de situatie is overal anders.

<br>

### **In grote steden**

In steden als Amsterdam en Rotterdam wonen heel veel mensen in appartementen. En zoals we net zagen hebben appartementen een hoger verbruik dan je zou verwachten.

Wat er gedaan zou kunnen worden:
- **Isolatie verbeteren** van appartementgebouwen. Dit lijkt ons het belangrijkst want dan dalen gas én elektriciteit allebei.
- **Stadsverwarming uitbreiden**. Nu is gemiddeld maar 15% van de woningen aangesloten, dat kan veel meer worden.
- **Bewoners informeren** over hoe ze energie kunnen besparen. Veel mensen weten dit gewoon niet.

<br>

### **In voorsteden en middelgrote gemeentes**

Hier is meer variatie in het type woningen. Je hebt hier rijtjeshuizen maar ook twee-onder-een-kappers.

Suggesties:
- **Bij nieuwbouw kiezen voor twee-onder-een-kap**. Uit onze data blijkt dit echt het efficientste woningtype te zijn.
- **Isolatiesubsidies** voor woningen van voor 2000. Dat is dus 83% van alle huizen, best veel.
- **Gezamelijke acties voor zonnepanelen**. Als een hele buurt meedoet is het goedkoper.

<br>

### **Op het platteland**

In dorpen staan relatief veel vrijstaande woningen. En die verbruiken dus meer.

Wat hier zou kunnen:
- **Prioriteit geven aan isolatie** van vrijstaande huizen. Die hebben grote oppervlaktes waardoor er veel warmte weglekt.
- **Warmtepompen stimuleren**. Die zijn goed geschikt voor grotere woningen.
- **Zonnepanelen**. Vrijstaande huizen hebben vaak grote daken dus dat is ideaal eigenlijk.

---

## **Het verschil is best groot**

Als je kijkt naar de gemiddelde verbruikscijfers per woningtype dan zie je duidelijke verschillen:

| Woningtype | Gemiddeld verbruik |
|------------|-------------------|
| Appartement | ~2200 kWh |
| Tussenwoning | ~2800 kWh |
| Hoekwoning | ~2970 kWh |
| Twee-onder-een-kap | ~3380 kWh |
| Vrijstaand | ~4125 kWh |

Vrijstaande woningen verbruiken dus bijna twee keer zoveel als appartementen.

Het kan opvallen dat appartementen de minste vebruik hebben. Dat klopt, maar dit is vooral omdat ze veel kleiner zijn. Wanneer je het verbruik per vierkante meter bekijkt en het gasverbruik meerekent, blijken twee-onder-een-kapwoningen efficiënter te zijn dan appartementen

<br>

## **Conclusie**

Dus eigenlijk komt het erop neer dat woningtype de belangrijkste factor is. En omdat gas en elektriciteit zo sterk samenhangen (die 0.81 correlatie) is isolatie echt de sleutel.

Samenvattend zouden we adviseren:
1. Focus op `isolatie`, vooral in appartementen en vrijstaande huizen
2. Bij nieuwbouw kiezen voor `twee-onder-een-kap` als het kan  
3. Extra aandacht voor huizen van voor `2000` (en dat is dus het merendeel)

De precieze aanpak verschilt per type gebied maar isolatie is overal de belangrijkste maatregel.
